```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = cv2.imread("../image/green003.png")
image_data = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)[:,:,2]
#image_data = image_data[32:180, 200:250] # red.png       #image_data[47:180, 294:339] # none.png
print("DEBUG| mean:",int(np.mean(image_data)))
#_,image_data = cv2.threshold(image_data,int(np.mean(image_data)),255,cv2.THRESH_BINARY)
print("DEBUG| shape rows:",image_data.shape[0], "columns:", image_data.shape[1])
midX = int(image_data.shape[1]/2)
midY = int(image_data.shape[0]/2)
thirdY = int(image_data.shape[0]/3)
rROI = ( int(thirdY/2) , midX ) # get the center point of each ROI
rROI = image_data[rROI[0]-5:rROI[0]+5, rROI[1]-5:rROI[1]+5]
yROI = ( midY, midX )
yROI = image_data[yROI[0]-5:yROI[0]+5, yROI[1]-5:yROI[1]+5]
gROI = ( midY+thirdY , midX )
gROI = image_data[gROI[0]-5:gROI[0]+5, gROI[1]-5:gROI[1]+5]
print("DEBUG| r:",int(np.mean(rROI)))
print("DEBUG| y:",int(np.mean(yROI)))
print("DEBUG| g:",int(np.mean(gROI)))

if int(np.mean(gROI)) > int(np.mean(yROI)) + int(np.mean(rROI)):
    print(">>> GREEN <<<")

elif int(np.mean(yROI)) > int(np.mean(gROI)) + int(np.mean(rROI)):
    print(">>> YELLOW <<<")

elif int(np.mean(rROI)) > int(np.mean(yROI)) + int(np.mean(gROI)):
    print(">>> RED <<<")

if int(np.mean(gROI)) > int(np.mean(yROI)) + int(np.mean(rROI)):
    print("GO!")
else:
    print("STOP!")
imgplot = plt.imshow(image_data, cmap='gray', vmin=0, vmax=255)
plt.show()
```
