```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# investigating OpenCV method used by team Stanley2.0

def red_filter(img):
    r = img[:,:,0]
    g = img[:,:,1]
    out = np.zeros_like(r)
    out[(r > 200) & (g <= 50)] = 1
    return out

img = cv2.imread("../image/TrafficLight_Img.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

imgplot = plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()

mask = red_filter(img)
imgplot = plt.imshow(mask, cmap='gray', vmin=0, vmax=1)
plt.show()

red_count = np.sum(mask)
if red_count > 10:
    print("red count: ",red_count)
    red_pos = np.where(mask==1)
    red_x = np.mean(red_pos[0])
    print(red_x)
    if red_x < 260 and red_x > 120:
        print(int(35-(red_x-200)/40))
    else:
        print("unknown")
```
